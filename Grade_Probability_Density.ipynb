# Calculating the probability density of student's grades

# Importing Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading Table

```python
grades = pd.read_excel('/content/drive/MyDrive/Student_Grades/Grades.xlsx')
```

```python
grades
```

# Caculate the mena of the "Grade" column

```python
mean_grade = grades['Grade'].mean()
mean_grade
```

# Calculate the standart deviation of the  "Grade" column

```python
std_dev_grade = grades['Grade'].std()
std_dev_grade
```

# Calculate the Probability Density for each studet

```python
grades['Probability Density'] = (1 / (std_dev_grade * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((grades['Grade'] - mean_grade) / std_dev_grade) ** 2)
```

# Create a new DataFrame with required columns

```python
probability_density = grades[['Student ID', 'Grade', 'Probability Density']]
```

# Display the new table

```python
probability_density
```

# Visualization

The curve reflects how the probability density varies across the different grades. We can notice that the 50% of the students will problably graded with 70.

```python
# @title Grade vs Probability Density

from matplotlib import pyplot as plt
probability_density.plot(kind='scatter', x='Grade', y='Probability Density', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

# Notes

Probability density it is a way of representing data distribution, is used used in continuous probability, where it represents the relative likehood of a variable taking a particular value within a range. In a normal distribution of grades, the probability density at a particular grade reflects how likely it is for a student's grade to fall near that value.
